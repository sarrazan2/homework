```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-12, 12, 1000)
y = np.sinc(x / np.pi)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)/x', color='dodgerblue', linewidth=2)

plt.xlim(-12, 12)
plt.ylim(-0.5, 1.2)

plt.xticks(np.arange(-12, 13, 1), rotation=45)
plt.yticks(np.arange(-0.5, 1.3, 0.1), rotation=45)

plt.grid(which='both', linestyle=':', linewidth=0.7, color = 'black')

plt.gca().set_facecolor('lightpink')

plt.title('График функции sin(x)/x', fontsize=16)
plt.xlabel('Значение x', fontsize=12)
plt.ylabel('sin(x)/x', fontsize=12)
plt.legend()
plt.show()
```
