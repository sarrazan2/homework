```python
import tkinter as tk
import random
import string

root = tk.Tk()
root.title("Генератор Паролей")

root.geometry("800x500")
root.configure(bg="lightblue")

def generate_password(length):
    password = ''
    letters = string.digits + string.punctuation + string.ascii_letters
    for _ in range(length):
        password += random.choice(letters)
    return password

def check_password(password):
    if len(password) >= 8 and any(i.isupper() for i in password) and any(i.islower() for i in password) and any(i.isdigit() for i in password) and any(i in string.punctuation for i in password):
        return "Надежный пароль", "green"
    elif all(i.isdigit() for i in password):
        return "Ненадежный пароль (только цифры)", "red"
    else:
        return "Ненадежный пароль", "red"
        
def strong_password():
    global password_label, length_n, n_label 
    def n_of_letters():
        global length, n_label, password_label 
        try:
            length = int(length_n.get())
            if length >= 8:
                password = generate_password(length)
                n_label.config(text="Введите желаемую длину пароля (не менее 8 символов):", font=("Arial", 12))
                n_label.pack(pady=10)
                password_label.config(text=f"Рекомендуемый пароль:\n{password}")
                password_label.pack()
            else:
                n_label.config(text="Недостаточно символов для создания надёжного пароля. Введите число больше или равно 8.")
        except ValueError:
            n_label.config(text="Введите целое число.")

    n_label = tk.Label(root, text="Введите желаемую длину пароля (не менее 8 символов):", font=("Arial", 12))
    n_label.pack(pady=10)
    length_n = tk.Entry(root)
    length_n.pack(pady=5)
    tk.Button(root, text="Подтвердить", command=n_of_letters).pack(pady=5)

def your_password():
    global result_label

    def checks_your_password():
        password = password_input.get()
        strength, color = check_password(password)
        result_label.config(text=strength, fg=color) 
        password_input.delete(0, tk.END)
        password_input.pack(pady=5) 
        
    password_label = tk.Label(root, text="Введите свой пароль:", font=("Arial", 12))
    password_label.pack(pady=10)
    password_input = tk.Entry(root, show="*")
    password_input.pack(pady=5)
    button2 = tk.Button(root, text="Проверить", command=checks_your_password)
    button2.pack(pady=5)
    result_label.pack(pady=10)


button1 = tk.Button(root, text="Сгенерировать надёжный пароль", command = strong_password, font=("Arial", 12))
button1.pack(pady=20)
button2 = tk.Button(root, text="Проверить ваш пароль", command = your_password, font=("Arial", 12))
button2.pack(pady=10)

password_label = tk.Label(root, text="", font=("Arial", 12))
result_label = tk.Label(root, text="", font=("Arial", 12))

button1.config(cursor="hand2")
button2.config(cursor="hand2")

root.mainloop()
```
